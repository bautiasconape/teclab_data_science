```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./AmesHousing.csv")
df.head()
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
test.isnull().sum()[test.isnull().sum() > 0].sort_values(ascending=False)
```

```python
corr = test.select_dtypes(include='number').corr()
corr_target = corr['SalePrice']

vars_relevantes = corr_target[abs(corr_target) > 0.6].index
corr_reducida = corr.loc[vars_relevantes, vars_relevantes]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_reducida)
plt.show()
```
